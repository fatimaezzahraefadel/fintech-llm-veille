# 01 — Exploration des données financières

Ce notebook explore les données collectées par le pipeline ETL.

```python
import pandas as pd
import plotly.express as px
from src.collectors.yahoo_finance import YahooFinanceCollector

collector = YahooFinanceCollector()
prices = collector.fetch_historical('AAPL')
df = pd.DataFrame([p.model_dump() for p in prices])
df.head()
```

```python
fig = px.line(df, x='date', y='close', title='Cours AAPL')
fig.show()
```
